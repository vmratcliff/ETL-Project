```python
#Strip leading and trailing spaces to allow for the reformatting to complete
Austin_crime["Date of Incident/Day of Week"] = Austin_crime["Date of Incident/Day of Week"].str.strip()
#Split the Date of Incident/Day of Week column out and change the format of the date

Austin_crime["Day of Week"]=Austin_crime["Date of Incident/Day of Week"].str.split("/", n=1, expand=True)[1]
Austin_crime["Date of Incident"]=Austin_crime["Date of Incident/Day of Week"].str.split("/", n=1, expand=True)[0]
Austin_crime["Date of Incident"] = Austin_crime["Date of Incident"].str.replace(" ","/")
Austin_crime
```

```python
#Alt option for Day of Week
Austin_crime['Date of Incident'] = pd.to_datetime(Austin_crime['Date of Incident'])
Austin_crime['Alt Day of Week'] = Austin_crime["Date of Incident"].dt.day_name()
Austin_crime
```

```python
#Combine the Victim totals and drop the break out columns
Austin_crime["Total Victims"] = Austin_crime["Number of vitims under 18"] + Austin_crime["Number of Victims Over 18"]
Austin_crime = Austin_crime.drop(['Number of vitims under 18', 'Number of Victims Over 18'], axis=1)
```

```python
#Combine the Offender totals and drop the break out columns
Austin_crime["Total Offenders"] = Austin_crime["Number of Offenders under 18"] + Austin_crime["Number of Offenders over 18"]
Austin_crime = Austin_crime.drop(['Number of Offenders under 18', 'Number of Offenders over 18'], axis=1)
Austin_crime
```
